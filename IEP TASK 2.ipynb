```python
df.drop(['employee_id','region'],axis = 1,inplace=True)      # Removes the 'employee_id' and 'region' columns from the DataFrame
```

```python
x= df.drop('is_promoted',axis=1)  # All features except 'is_promoted'
y=df['is_promoted']  # Target variable
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_department = LabelEncoder()  # Create a label encoder for the 'department' column
le_education = LabelEncoder()   # Create a label encoder for the 'education' column
le_gender = LabelEncoder()   # Create a label encoder for the 'gender' column
le_recruitment_channel = LabelEncoder() # Create a label encoder for the 'recruitment_channel' column
```

```python
x['department']=le_department.fit_transform(x['department'])   # Convert 'department' categories to numbers
x['education']=le_education.fit_transform(x['education'])       # Convert 'education' categories to numbers
x['gender']=le_gender.fit_transform(x['gender'])               # Convert 'gender' categories to numbers
x['recruitment_channel']=le_recruitment_channel.fit_transform(x['recruitment_channel']) # Convert 'recruitment_channel' categories to numbers
```

```python
# correlation heatmap 
plt.figure(figsize=(10,6))
sns.heatmap(x.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split # Import function to split data into training and test sets
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42) # Split the data into training (80%) and testing (20%) sets
```

```python
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier(n_estimators=100)  # Create a RandomForestClassifier model

# Train model
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train) 

rfc.score(x_test,y_test)*100
```

```python
from sklearn.linear_model import LogisticRegression  # Import the Logistic Regression model from scikit-learn
```

```python
lr = LogisticRegression()  # Create a Logistic Regression model
lr.fit(x_train,y_train) # Train the model using the training data
```

```python
lr.score(x_test,y_test)*100 # Shows the model's accuracy on the test set as a percentage
```

```python
from sklearn.tree import DecisionTreeClassifier  
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
dt_model = DecisionTreeClassifier()    # Create a Decision Tree model
dt_model.fit(x_train, y_train)         # Train the model on the training data
```

```python
za = DecisionTreeClassifier() # Create a DecisionTreeClassifier model
za.fit(x_train,y_train)  # Train the model using the training data
```

```python
za.score(x_test,y_test)*100  # Shows the model's accuracy on the test set as a percentage
```

```python
from sklearn.metrics import accuracy_score, f1_score

# Logistic Regression
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_acc = accuracy_score(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_test)
dt_acc = accuracy_score(y_test, dt_pred)
dt_f1 = f1_score(y_test, dt_pred)

# Random Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
rf_acc = accuracy_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred)

# Print results
print("Logistic Regression - Accuracy:", lr_acc, "F1 Score:", lr_f1)
print("Decision Tree        - Accuracy:", dt_acc, "F1 Score:", dt_f1)
print("Random Forest        - Accuracy:", rf_acc, "F1 Score:", rf_f1)
```
